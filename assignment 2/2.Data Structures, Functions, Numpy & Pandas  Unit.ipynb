```python
#1 .Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.
#creating the list of 5  integers

a = [5,12,23,45,67,86]
print(type(a))
print("1.The list is :",a)

#append
a.append(5)
print("2.the list after appending the value 5:",a) 

#extend
a.extend('vis')
print("3.the list after extending (vis) :",a)

#insert
a.insert(5,10)
print("4.The list after inserting (5,10):",a)

#remove
a.remove(a[1])
print("5.the list after removing index 1:",a)

#pop
pop = a.pop(2)
print(f"pop value is :",pop)
print("6.the list after pop the 2 index: ",a)

#index
index = a.index('v')
print(f"7.the index value of v:",index)

#count
count = a.count(5)
print(f"8.the number of 5 available in a list is  :" ,count)

#reverse
a.reverse()
print("9.The reverse of list is :",a)

#sort
b= [x for x in a if type(x) == int or type(x) == float]  #here we are taking only int and float so the string variable will be removed
b.sort()
print("10.The sorted list is :",b)

#clear
a.clear()
print("11.clearing the list :",a)
```

- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'

```python
marks = int(input("Enter the mark :"))

if marks >= 90 :
    print(f"The Grade of {marks} is A")
elif marks >= 75 and  marks <=89:
    print(f"The Grade of {marks} is B")
elif marks >= 50 and marks <=74:
    print(f"The Grade of {marks} is c")
else:
    print(f"The mark is {marks} and grade is fail")
     
```

7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
num = int(input("Enter the number :"))

if(num>0 and num%2 == 0):
    print(f"the number ({num}) is positive and even")
elif(num>0 and num%2 == 1):
    print(f"the number ({num}) is positive and odd")
else:
    print(f"the number ({num}) is zero or negative")
```
