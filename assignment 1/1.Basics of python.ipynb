EXERCISE 1 :

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
num = int(input("Enter the number : "))

if num > 1:
    for i in range(2,num):
        if num % i == 0:
            print(num ," is not a prime number..")
            break
        else:
            print(num , " is Prime Number...")
            break
            
else:
    print(num ," is not a Prime Number...")
            

```

Exercise 2: Product of Random Numbers
Develop a Python program. that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
import random 

a = random.randint(1,10)
print(a)

b = random.randint(1,10)
print(b)

print("what is the product of " ,a ,"and",b)
answer = int(input("Enter the product :"))

if answer == a*b:
    print("Correct.............")
else:
    print("The correct answer is : " , a*b)
```

Exercise 3 : Squares of Even/Odd Numbers

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.


```python
choice  = input("enter the choice even or odd to perform square :")

if choice.lower() == "even":
    print("The  choice choosen is even...")
    for i in range(100,201):
        if i % 2 == 0:
            print("even numbers :",i,"and square of even numbers :",i**2)
        
elif choice.lower() == "odd":
    print("The choice choosen  is odd...")
    for i in range(100,201):
        if i % 2 == 1:
            print("odd numbers :",i,"and square of odd numbers :",i**2)
else:
    print("The given Choice is Invalid...")
        
```

Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400



```python
Room_charge_per_day = 2000
Doctor_consultant_fee = 1500
Lab_test_charge = 300

no_of_days = 4
lab_test = 3
medicine_charge = 2400

room_bill = Room_charge_per_day*no_of_days
lab_bill = Lab_test_charge*lab_test

total_bill_before_discount = room_bill + lab_bill + Doctor_consultant_fee + medicine_charge

if total_bill_before_discount > 10000:
    discount = total_bill_before_discount * 0.10
else:
    discount = 0
    
final_bill = total_bill_before_discount - discount

print(f"room bill amount : RS.{room_bill}")
print(f"doctor consultant fee : RS.{Doctor_consultant_fee}")
print(f"lab bill amount : RS.{lab_bill}")
print(f"medical bill amount : RS.{medicine_charge}")
print("------------------------------------------------------")
print(f"total bill before discount amount  : RS.{total_bill_before_discount}")
print(f"Discount : RS.{discount}")
print("------------------------------------------------------")
print(f"Final bill amount : RS.{final_bill}")
```
